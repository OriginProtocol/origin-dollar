```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(filename):
    base = pd.read_csv(filename)
    for x in base:
        if " " in x or "action" in x:
            continue
        else:
            base[x] = base[x].apply(int)
    base['pre_mix'] = base['pre_pool_0'] / (base['pre_pool_0'] + base['pre_pool_1'])
    base['before_mix'] = base['before_pool_0'] / (base['before_pool_0'] + base['before_pool_1'])
    base['after_mix'] = base['after_pool_0'] / (base['after_pool_0'] + base['after_pool_1'])
    base['before_profit'] = base['before_vault'] - base['pre_vault']
    base['after_profit'] = base['after_vault'] - base['before_vault']
    return base

base = load_data("deposit_stats.csv")
balance_base = load_data("balance_stats.csv")
withdraw_base = load_data("withdraw_stats.csv")
withdrawall_base = load_data("withdrawall_stats.csv")
```

```python
df = base
plt.axhline(0, c="grey", linewidth=0.5)
plt.axhline(0, c="black", linewidth=0.4)
for before_mix, rows in df.groupby(df['before_mix']):
    plt.plot(rows['action_mix'], rows['after_profit'])
```

```python
df = base
plt.title("Deposit profit")
plt.axhline(0, c="grey", linewidth=0.5)
plt.axhline(0, c="black", linewidth=0.4)
for before_mix, rows in df.groupby(df['before_mix']):
    plt.plot(rows['action_mix']*100, rows['after_profit'], label="{:.1f}%".format(100*float(before_mix)))
plt.ylim([-1e18,1e18])
plt.xlabel("Deposit Mix")
plt.ylabel("Deposit Profit")
plt.legend()
```

```python
df = base
df = df.sort_values('before_mix')
plt.plot(df['before_mix'],df['before_profit'])
```
